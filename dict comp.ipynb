```python
nums = [1,2,3,4]
```

```python
# basic 

{x:x**2 for x in nums}
```

```python
# with if

{x:x**2 for x in nums if x%2!=0}
```

```python
# with if-else

{x:'even' if x%2==0 else 'odd' for x in nums}
```

```python
# form 2 lists

keys = [1,2,3]
vals = ['one','two','three']

{k:v for k,v in zip(keys,vals)}
```

```python
# swap key-value

d = {1: 'one', 2: 'two', 3: 'three'}

{v:k for k,v in d.items()}
```

```python
# nested loops

{(x,y) : x*y for x in range(2,4) for y in range(2,4)}
```
