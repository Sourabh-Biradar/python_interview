```python
# args & kwargs
# args : vars packed as tuple
# kwargs : dict

def person(*args,**kwargs):
    print(args)
    print(kwargs)

person(3,4,5.1,name='Abc',age=33)
```

```python
def person(*l,**info):
    print(l)
    print(info)

person(3,4,5.1,name='Abc',age=33)
```

```python
# lambda func

res = lambda x,y,z : x+y*z

res(3,4,5)
```

```python
# variable scope
# global keyword

x = 'global'

def fn():
    x = 'local'

fn()
print(x)

def update():
    global x
    x = 'GLOBAL'

update()
print(x)
```

```python
# enclosed
# nonlocal keyword

def outer():
    y = 10

    def inner():
        nonlocal y
        y = 11

    inner()
    print(y)

outer()
```

```python
# multiple returns
# returns as tuple

def fn():
    return 3,5.0,'Seven'

print(fn())
```

```python
# recursion function

def fact(n):
    if n==0 : return 1

    return n * fact(n-1)

print(fact(5))
```
