```python
words = ['apple', 'banana', 'kiwi']

sorted(words,key=len)
```

```python
# dict sorting : list of dicts NOT directly dict

marks = [
    {'name':'Abc','score':34},
    {'name':'Lmn','score':78},
    {'name':'Xyz','score':61},
]

sorted(marks,key=lambda x:x['score'],reverse=True)
```

```python
# sorting tuple

marks = [("John", 85), ("Sara", 92), ("Alex", 78)]
sorted(marks, key=lambda x: x[1])
```

```python
# searching list 

l = [12,32,5,9]

print(l.index(32))
print(5 in l)
```

```python
# searching dict

d = {'name':'Xxx','age':22}

print(d['age'])
print(23 in d.values())
```

```python
# searching Set

s = {1,7,4}

7 in s
```

```python
# list of dicts
# accessing

l = [
    {'name':'one','age':1},
    {'name':'two','age':2},
    {'name':'three','age':3},
]

for i in l:
    print(i)
    print(i['name'],i['age'])
```
