Basic 

with if

with if-else : if-esle ; loop

with nested loop : 2 vars ; loop & loop

apply function : function ; loop

flattern : [x for row in m for x in row]

```python
nums = [1,2,3,4]
```

```python
# basic 
[x * 2 for x in nums]
```

```python
# with if

[x for x in nums if x%2==0]
```

```python
# with if-esle

[x if x>2 else 0 for x in nums]
```

```python
['even' if x%2==0 else 'odd' for x in nums]
```

```python
# nested loop

[(x,y) for x in nums for y in nums]
```

```python
# apply fucntion

[float(x) for x in nums]
```

```python
# flattern
m = [[2,4,6],[1,3,5]]

[x for row in m for x in row]
```
